centrality of karate club network

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
#import pprint

G = nx.karate_club_graph()
plt.figure(figsize=(5, 5))
nx.draw_spring(G, node_size=400, node_color='red', with_labels=True, font_weight='bold')
#print("degree centrality:",list(nx.degree_centrality(G).values()))
v = list(nx.degree_centrality(G).values())
s = ("     degree centrality: "+', '.join(['%.2f']*len(v))) % tuple(v)
print(s, "biggest:", np.argmax(v))
#print("betweenness centrality",nx.betweenness_centrality(G))
v = list(nx.betweenness_centrality(G).values())
s = ("betweenness centrality: "+', '.join(['%.2f']*len(v))) % tuple(v)
print(s, "biggest:", np.argmax(v))
#print("closeness centrality",nx.closeness_centrality(G))
v = list(nx.closeness_centrality(G).values())
s = ("  closeness centrality: "+', '.join(['%.2f']*len(v))) % tuple(v)
print(s, "biggest:", np.argmax(v))
#print("eigenvector centrality",nx.eigenvector_centrality(G))
v = list(nx.eigenvector_centrality(G).values())
s = ("eigenvector centrality: "+', '.join(['%.2f']*len(v))) % tuple(v)
print(s, "biggest:", np.argmax(v))
#print("PageRank centrality",nx.pagerank(G))
v = list(nx.pagerank(G).values())
s = ("   PageRank centrality: "+', '.join(['%.2f']*len(v))) % tuple(v)
print(s, "biggest:", np.argmax(v))
#print("Katz centrality",nx.katz_centrality(G))
v = list(nx.katz_centrality(G).values())
s = ("       Katz centrality: "+', '.join(['%.2f']*len(v))) % tuple(v)
print(s, "biggest:", np.argmax(v))
```

hubs and autorities

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import numpy.linalg as LA 

G = nx.DiGraph()
G.add_nodes_from(range(0,5))
G.add_edges_from([(1,0),(0,2),(2,4),(4,2),(0,4),(4,3),(5,4)])
plt.figure(figsize=(5, 5))
nx.draw_spring(G, node_size=400, node_color='red', with_labels=True, font_weight='bold')
h,a=nx.hits(G)
h = list(h.values())
s = ("      hub : "+', '.join(['%.2f']*len(h))) % tuple(h)
print(s, "biggest:", np.argmax(h))
a = list(a.values())
s = ("authority : "+', '.join(['%.2f']*len(a))) % tuple(a)
print(s, "biggest:", np.argmax(a))
```

transitivity, local clustering coefficient

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

G = nx.Graph()
G.add_nodes_from(range(0,4))
G.add_edges_from([(0,1),(1,2),(1,3),(1,4),(2,3),(3,4),(2,4)])
plt.figure(figsize=(5, 5))
nx.draw_spring(G, node_size=400, node_color='red', with_labels=True, font_weight='bold')
print("transitivity (built in function):", nx.transitivity(G))

A = nx.adjacency_matrix(G).todense()
#A = np.array(A, dtype = np.float64)
print(nx.triangles(G))
print("number of triangles:",int(sum(nx.triangles(G).values())/3))
A2 = A*A-np.diag(np.diag(A*A))
print("number of triads:",np.sum(A2)/2)
print("transitivity :",sum(nx.triangles(G).values())/3*3/(np.sum(A2)/2))

print("average of local clustering coefficient (built in function):", nx.average_clustering(G))
print("local clustering coefficient:", nx.clustering(G))
print("local clustering coefficient:", sum(nx.clustering(G).values())/nx.number_of_nodes(G))
```

signed network

```python
import networkx as nx
import numpy as np
G = nx.Graph()
G.add_edges_from([(1,2),(2,3),(1,3),(2,4)], relationship=1)
G.add_edges_from([(3,4)], relationship=-1)
labels = {e[0:2]:e[2]['relationship'] for e in G.edges(data=True)} 
plt.figure(figsize=(5, 5))
layout = nx.spring_layout(G)
nx.draw(G,pos=layout, with_labels=True, node_size=300)
nx.draw_networkx_edge_labels(G,pos=layout, edge_labels=labels,font_size=15)

#print(G.edges())
print(G.edges(data=True))
#{print(e[2]['relationship']) for e in G.edges(data=True)}

triangles = [c for c in nx.cycle_basis(G) if len(c)==3]
triangle_types={}
for triangle in triangles:
    tri=nx.subgraph(G,triangle)
    #take the product of the edge relationships. If there are an odd number of -1s, the triangle is unbalanced.
    triangle_types[tuple(tri.nodes())]=np.product([x[2]['relationship'] for x in tri.edges(data=True)])
print(triangle_types)
```

cosine similarity, Pearson correlation coefficient

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import numpy.linalg as LA 

G = nx.complete_graph(5)
plt.figure(figsize=(4, 4))
nx.draw_circular(G, node_size=300, node_color='red', with_labels=True, font_weight='bold')
A = nx.adjacency_matrix(G).todense()
A = np.array(A, dtype = np.float64)
print(A)
print("cosine similarty (0,1):",np.dot(A[0],A[1])/(LA.norm(A[0])*LA.norm(A[1])))

print("Pearson correlation coefficient (0,1):", np.cov(A,bias=1)[0][1]/(np.std(A[0])*np.std(A[1])))
```

modularity

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
from networkx.algorithms import community

G = nx.Graph()
G.add_nodes_from(range(0,5))
G.add_edges_from([(0,1),(1,2),(0,2),(2,3),(3,4),(4,5),(3,5)])
plt.figure(figsize=(5, 5))
nx.draw_spring(G, node_size=400, node_color='red', with_labels=True, font_weight='bold')

# partition for maximizing modularity
c = list(community.greedy_modularity_communities(G))
print(c)
mod = community.modularity(G, c)
print(mod)

# another partition
c2 = [{0, 2, 4}, {1, 3, 5}]
print(c2)
mod2 = community.modularity(G, c2)
print(mod2)
```
